```python
#import all the required libraries
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,TimeoutException
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)
```

```python
url = "https://www.olx.in/items/q-cars"
driver.get(url)
time.sleep(10)
```

```python
url = []
urls = driver.find_elements_by_xpath("//li[@class='EIR5N']/a")
for i in urls:
    url.append(i.get_attribute('href'))
```

```python
brand =[]
fuel = []
no_owner=[]
location=[]
driven_kilometers =[]
Transmission=[]
price=[]
history =[]
```

```python
for i in url:
    driver.get(i)
    try:
        brand.append(driver.find_element_by_xpath("//div[@class='_35xN1']").text)
    except NoSuchElementException:
        brand.append('---')
    try:
        fuel.append(driver.find_element_by_xpath("//div[@class='_3qDp0']").text)
    except NoSuchElementException:
        fuel.append('---') 
    try:
        no_owner.append(driver.find_element_by_xpath("//div[@class='_1gasz']").text)
    except NoSuchElementException:
        no_owner.append('---')
    try:
        location.append(driver.find_element_by_xpath("//div[@class='_1l939']/div[2]").text)
    except NoSuchElementException:
        location.append('---')
    try:
        driven_kilometers.append(driver.find_element_by_xpath("//div[@class='aOxkz']/div[2]").text)
    except NoSuchElementException:
        driven_kilometers.append('---')
    try:
        Transmission.append(driver.find_element_by_xpath("//div[@class='aOxkz']/div[3]").text)
    except NoSuchElementException:
        Transmission.append('---')
    try:
        price.append(driver.find_element_by_xpath("//div[@class='_3FkyT']").text)
    except NoSuchElementException:
        price.append('---')
    
```

```python
olx = pd.DataFrame({'brand':brand,'fuel':fuel,'no_owner':no_owner,'location':location,'driven_kilometers':driven_kilometers,'Transmission':Transmission,'price':price})
```
